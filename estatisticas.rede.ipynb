#### Análise de Engajamento em Redes Sociais — Estatística para Criadores de Conteúdo

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = {
    'Postagem': ['Look do dia', 'Dica de beleza', 'Reels de humor', 'Receita fit', 'Frase motiv.'],
    'Curtidas': [320, 410, 890, 510, 300],
    'Comentários': [45, 60, 120, 70, 40],
    'Compartilhamentos': [12, 25, 80, 30, 10]
}
```

```python
df = pd.DataFrame(dados)
```

```python
print(" Engajamento por postagem:")
print(df)
```

```python
media_curtidas = df['Curtidas'].mean()
media_comentarios = df['Comentários'].mean() 
media_compartilhamentos = df['Compartilhamentos'].mean() 
```

```python
print(f"\n Média de curtidas: {media_curtidas:.2f}")
print(f" Média de comentários: {media_comentarios:.2f}")
print(f" Média de compartilhamentos: {media_compartilhamentos:.2f}")
```

```python
df['Engajamento Total'] = df['Curtidas'] + df['Comentários'] + df['Compartilhamentos']
print("\n Postagem com maior engajamento total:")
print(df.loc[df['Engajamento Total'].idxmax()])
```

```python
plt.figure(figsize=(10, 6))
df.set_index('Postagem')[['Curtidas', 'Comentários', 'Compartilhamentos']].plot(kind='bar', figsize=(10, 6))
plt.title(' Engajamento por tipo de conteúdo')
plt.ylabel('Quantidade')
plt.xlabel('Postagem')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("\n Interpretação:")
print("O Reels de humor teve o maior engajamento total, indicando que conteúdos leves e divertidos geram mais interação.")
print("A correlação positiva entre curtidas e compartilhamentos sugere que postagens populares tendem a ser mais compartilhadas.")
```
